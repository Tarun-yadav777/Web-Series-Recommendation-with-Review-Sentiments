```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('All_Streaming_Shows.csv')
```

```python
plt.figure(figsize=(16,6))
plt.xticks(rotation=90)
plt.title("No. Of Web Series over years.(1901-2020)")
sns.countplot(x=df['Year Released'])
```

```python
last_decade = sum([y for y in df['Year Released'].value_counts()][:10])
rest = sum([y for y in df['Year Released'].value_counts()][10:])
```

```python
labels = ['last_decade', 'rest']
sizes = [last_decade, rest]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 
plt.title('% of Web Serie Released in last decade and rest')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Distribution of IMDB Ratings")
sns.distplot(df['IMDB Rating'], bins=40)
```

```python
plt.figure(figsize=(12,6))
plt.title('Distribution of R Ratings')
sns.distplot(df['R Rating'], bins=40)
```

```python
plt.figure(figsize=(12,6))
season_no = df['No of Seasons'].apply(lambda x: x.split("S")[0])
plt.title('Distribution of No of Seasons')
sns.distplot(season_no, bins=40)
```
